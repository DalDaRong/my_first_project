```python
result=[]
#페이지를 1부터 2페이지까지 반복
#for page in range(1,2)#한페이지만 우선 보자
for page in range(1,57+1):#57페이지까지 보자
    #이제 주소를 알아야함# f"{}"를 사용해 range를 저 아래에 페이지 넘버로 치환해줘요....
    url=f"https://www.hollys.co.kr/store/korea/korStore.do?pageNo={page}&sido=&gugun=&store="
    #print(url)
    html=urllib.request.urlopen(url)#소스 코드 가져오고 오픈. 근데 이 상태에서는 뭘 못찾지.
    soup=BeautifulSoup(html,"html.parser") #코드 찾기 위해서는 뷰티풀숲으로 숲 객체 만들어 파싱하기.:문법 규칙에 따른 문자열을, 해당 문법을 바탕으로 단어의 의미나 구조를 분석하는 것을 의미
    #print(soup)
    #<tbody>가 <tr>들을 감싸고 있네. tbody로 찾아야겠다.
    tbody=soup.find("tbody")
    #print(tbody)
    #여기서 한 지점만 보고싶은뎅... 여기서 <tr>로 한 줄씩 10번 묶여있었지! 여러개 찾으려면  find.all
    trs=tbody.find_all("tr")
    #print(trs) #오 리스트로 잘 묶였어. 0번째 1번째로 볼 수 있겠다.
    #print(trs[1])
    for store in trs:
        tds=store.find_all("td") #td들로 조각조각 찾아보고싶으니 데이터로 쪼갤래'
        print(tds[0].get_text(),end="/")#지역
        print(tds[1].get_text(),end="/")#지점
        #print(tds[1].get_text())는 필요없는 정보니까 건너뛸래
        print(tds[3].get_text(),end="/")#주소
        print(tds[5].get_text())#전화번호 오옹 데이터 하나씩 잘나온다...!
        #이제 반복문으로 돌려볼까.
        print()
        #이 나온 결과 값을 result에 누적시키자.
        result.append([tds[0].get_text()]+
                      [tds[1].get_text()]+
                      [tds[3].get_text()]+
                      [tds[5].get_text()])
        
        #자 이걸 페이지 단위로 해보자!!! 위에 페이지 범위 바꿔!!!
        
        
        
    
    
```

```python
len(result)
```

```python
import pandas as pd
col_name=["지역","지점명","주소","전화번호"]
df_hollys=pd.DataFrame(result,columns=col_name)
df_hollys.to_csv("hollys지점정보.csv",encoding="cp949")
df_hollys
```

```python
pattern=re.compile("(\d{6})[-]\d{7}") 
# 6자리 숫자[-]7자리 숫자.
# \d는 숫자를 의미한다. [0-9]와 동일하다.
# \D는 숫자가 아닌 것.[^0-9]와 동일하다.
# \s는 공백문자. (space, tab, 줄바꿈표시 등)
# \S는 공백이 아닌 것.
# \w는 숫자와 문자. [a-zA-Z0-9]
# \W는 숫자/문자 아닌것,[^a-zA-Z0-9]
print(pattern.sub("\g<1>-*******",data))
```
