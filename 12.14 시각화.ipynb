```python
import matplotlib
matplotlib.rcParams["font.family"]="Malgun Gothic"

city=["서울","인천","대전","대구","울산","부산","광주"] #위도 lat 경도 long 표시하기
lat=[37.56,37.45,36.35,35.87,35.53,35.18,35.16]
long=[126.97,126.70,127.38,128.6,129.31,129.07,126.85]
#pop_den 인구밀도
pop_den=[16154,2751,2839,2790,1099,4454,2995]

size=np.array(pop_den)*0.2 # 인구 밀도에 따라 커지고 작아지는 점
colors=['r','g','b','c','m','k','y']
plt.scatter(long,lat,s=size,c=colors,alpha=0.5)
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.title("population(2017)")
for x,y,name in zip(long, lat, city): #x는 경도에서 y점은 위도에서 이름은 city에서
    plt.text(x,y,name)
plt.show
```

### 운동 향상 시각화

```python
#313
member_IDs=["m_01","m_02","m_03","m_04"]
before_ex=[27,35,40,33] #한달 전 운동양
after_ex=[30,38,42,37]  #이번 달 운동양
```

```python
matplotlib.rcParams["axes.unicode_minus"]=False
n_data=len(member_IDs)
index=np.arange(n_data)#데이터 더 많아져도 넘파이화 해놓으면 일일이 추가 안해도 됨.
plt.bar(index,before_ex)
plt.show()
```

```python

n_data=len(member_IDs)
index=np.arange(n_data)
plt.bar(index,before_ex,width=0.3) #막대의 폭
plt.text(0,27,"27")
plt.show()
```

```python

n_data=len(member_IDs)
index=np.arange(n_data)
plt.bar(index,before_ex,width=0.3,tick_label=member_IDs,color=colors) #tick 라벨 밑에 이름.
plt.text(0,27,"27")
plt.show()
```

```python
plt.barh(index,before_ex,height=0.3,tick_label=member_IDs,color=colors) #height 가로 폭
plt.text(27,0,"27") #가로니까 x y 값 위치 바뀜
plt.show()
```

```python
plt.bar(index,before_ex,width=0.3,color="b", label="before")
plt.bar(index+0.3,after_ex,width=0.3,color="g",label="after") #옆으로 0.3만큼 옮기기

plt.xticks(index+0.15,member_IDs) #tick 아래에 적어주기.
plt.legend()#라벨 위에 보여주기
plt.show()
```

## add chart 엑셀파일을 추가하기

```python
excel_writer=pd.ExcelWriter("시간대별생산량.xlsx",engine="xlsxwriter")
df.to_excel(excel_writer,index=True, sheet_name="Sheet1")
workbook=excel_writer.book#워크북 만들기
worksheet=excel_writer.sheets["Sheet1"]

#차트 만들기
chart=workbook.add_chart({"type":"line"}) #라인 타입의 차트를 만들겠다
chart.add_series({"values":"=sheet1!$B$2:$B$8"}) #sheet1 B2에서부터 B8까지
chart.add_series({"values":"=sheet1!$C$2:$C$8"}) #sheet1 C2에서부터 C8
#worksheet.insert_image('D2',"시간대별생산량.png")#D2셀에 그림 파일 삽입.
worksheet.insert_chart("D2",chart)#1행 3열=D2
excel_writer.save()

```
