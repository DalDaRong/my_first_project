## p438 웹에서 이미지 가져오기


```python
import requests
url="https://www.python.org/static/img/python-logo@2x.png"
html_image=requests.get(url)
html_image
```

```python
import os
file_name=os.path.basename(url)
#basename은 여기서 python-logo
file_name
```

```python
image_file=open(file_name,"wb") #wb메모장에서 안열림 w열림
#조각조각내서 저장할래
chunk_size=1000000 #백만바이트씩 조각낼래
for chunk in html_image.iter_content(chunk_size):
    image_file.write(chunk)
image_file.close() #다 끝났으면 열었던 파일 닫아.
```

## 예외처리문 try, except, pass

```python
#문제가 생겼을 때 문제에 대처하는 방법. 에러를 고치거나, 에러 일어나는 부분은 넘어가거나.
```

```python
분모=[1,2,4,6,2,9,0,10,20,2]
분자=100
for num in 분모:

    print(분자/num)
    
```

```python
#division by zero 뜻은 0으로 나누려 했으니 안된다는 말이다.
```

```python
분모=[1,2,4,6,2,9,0,10,20,2]
분자=100
for num in 분모:
    if num==0:
        continue
    print(분자/num)
#이렇게 고쳐주는건 옛날처리방식. 알고리즘에 대한 완전한 이해가 필요한 해결법
```

```python
분모=[1,2,4,6,2,9,0,10,20,2]
분자=100
for num in 분모:
    try:
        print(분자/num)
    except:
        pass #문제 생겨도 넘어가라는 명령어.
        print("문제생겼어요.")
    
```

```python
### 웹에서 사진 누락된 부분이 있어도, 예외처리문 사용해서 에러 없이 실행해보자.
url="https://www.reshot.com/free-stock-photos/animals/"
image_html=requests.get(url).text # 컨텐츠만 갖고오기
soup=BeautifulSoup(image_html,"html.parser")
# 그 중에서 div 골라라. 
res=soup.find("div",{"class":"item-grid-masonry"})#속성: 클래스 item-grid-masonry에서 div 찾아라
images=res.find_all("img")
for i in range(len(images)):
    try:
        url=images[i].get("src")
        
        html_image=requests.get(url)
        file_name=os.path.basename(url)
        file_name="reshot/"+file_name
        
        image_file=open(file_name,"wb") #wb메모장에서 안열림, 이 이름으로 된 파일 열자
        chunk_size=1000000
        for chunk in html_image.iter_content(chunk_size):
            image_file.write(chunk)
        image_file.close()
    except:
        pass
```
